```python
from langgraph.graph import StateGraph
from typing import TypedDict, Dict
```

```python
class AgentState(TypedDict): # shared data structure for the agent's state that keep track of the message
    message: str

def personalized_greeting_node(state: AgentState) -> AgentState:
    """A node that adds a personalized greeting message to the state."""

    state['message'] = f"{state['message']}, you're doing an amazing job learning LangGraph!"

    return state
```

```python
graph = StateGraph(AgentState)
graph.add_node("personalized_greeter", personalized_greeting_node)

graph.set_entry_point("personalized_greeter")
graph.set_finish_point("personalized_greeter")
app = graph.compile()
```
